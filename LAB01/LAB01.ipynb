```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data Loading
df = pd.read_csv('cancer patient data sets.csv')

# Unique Categories in 'Level'
unique_levels = df['Level'].unique()
print("Unique Levels:", unique_levels)

# Data Overview
print("Total columns:", df.shape[1])
print("First 50 records:\n", df.head(50))
print("Last 30 records:\n", df.tail(30))
df.info()
print("Dimension:", df.shape)
print("Size:", df.size)

# Columns and Samples
print("Number of columns:", df.shape[1])
print("Number of samples:", df.shape[0])

# Category Counts
print("Value counts for 'Level':\n", df['Level'].value_counts())
sns.countplot(x='Level', data=df)
plt.title('Counts of Level Categories')
plt.xlabel('Level')
plt.ylabel('Count')
plt.show()

# Extract Rows 550-900 (Indices 549-899)
extracted_g = df.loc[549:899, 'Gender':'Coughing of Blood']
extracted_g.to_csv('extracted_g.csv', index=False)
print("Statistical Summary for Extracted Data (G):\n", extracted_g.describe())
print("Median:\n", extracted_g.median())
print("Mode:\n", extracted_g.mode().iloc[0])

# Extract Last 20 Rows for Specific Columns
col_start = df.columns.get_loc('Shortness of Breath')
col_end = df.columns.get_loc('Frequent Cold') + 1
extracted_h = df.iloc[-20:, col_start:col_end]
extracted_h.to_csv('extracted_h.csv', index=False)
print("Statistical Summary for Extracted Data (H):\n", extracted_h.describe())
print("Median:\n", extracted_h.median())
print("Mode:\n", extracted_h.mode().iloc[0])

# Extract Specific Rows and Calculate Average Alcohol Use
rows = [109, 219, 359, 439, 655, 777, 201, 888]
extracted_i = df.iloc[rows]
avg_alcohol = extracted_i['Alcohol use'].mean()
print("Average Alcohol use:", avg_alcohol)
```

```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data Loading
df = pd.read_csv('cancer patient data sets.csv')

# Histogram of Age
df['Age'].hist(bins=20)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot of Age vs Obesity
plt.scatter(df['Age'], df['Obesity'])
plt.title('Scatter of Age vs Obesity')
plt.xlabel('Age')
plt.ylabel('Obesity')
plt.show()

# Distribution of Age
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

# Boxplot of Obesity
sns.boxplot(y=df['Obesity'])
plt.title('Boxplot of Obesity')
plt.ylabel('Obesity')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Distribution of Frequent Cold
sns.histplot(df['Frequent Cold'], kde=True)
plt.title('Distribution of Frequent Cold')
plt.xlabel('Frequent Cold')
plt.ylabel('Density')
plt.show()

# Boxplot of Obesity for Outliers
sns.boxplot(y=df['Obesity'])
plt.title('Boxplot of Obesity to Detect Outliers')
plt.ylabel('Obesity')
plt.show()
```
